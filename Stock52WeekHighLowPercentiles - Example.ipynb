```python
tickers = random.sample(large_cap, 10) + random.sample(mid_cap, 10) + random.sample(small_cap, 10) 

# Set the ticker, start date, and end date
start_date = '2000-01-01'
end_date = '2023-12-01'

percentiles = []
for ticker in tickers: 
    # Create an instance of the Stock52WeekHighLowPercentiles class
    stock_data = Stock52WeekHighLowPercentiles(ticker, start_date, end_date)

    # Calculate the 52 week high and low percentiles 
    data = stock_data.calculate_52week_percentiles()
    if ticker in large_cap:
        data['size'] = 'Large Cap'
    elif ticker in mid_cap:
        data['size'] = 'Mid Cap'
    elif ticker in small_cap:
        data['size'] = 'Small Cap'  
    
    percentiles.append(data)
        
    

# Print the percentiles
df = pd.concat(percentiles)
```

```python
df.reset_index(inplace=True)
```

```python
data = df.groupby(['Date','size']).agg({'52wk_percentile_21_sma': 'mean'}).reset_index()
data.head()
```

```python
#Set size of figure
plt.figure(figsize=(20,8))

#Create a list of tickers
sizes = data['size'].unique()

#Loop through each ticker and plot 52wk_percentile line
for t in sizes:
    images = data[data['size'] == t].copy()
    plt.plot(images['Date'], images['52wk_percentile_21_sma'], label=t)

#Set labels
plt.xlabel('Index')
plt.ylabel('52wk_percentile_21_sma')

#Set figure size
plt.figsize=(20,12)


#Show legend
plt.legend()

#Show plot
plt.show()
```
